```python
#Averaging
kernel = np.ones((5,5), np.float32)/25
dst = cv2.filter2D(gsc_img, -1, kernel)

plt.subplot(121), plt.imshow(gsc_img), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Gaussian Blur
kernel = np.ones((5,5), np.float32)/25
blur = cv2.GaussianBlur(gsc_img, (5,5), cv2.BORDER_DEFAULT)

plt.subplot(121), plt.imshow(gsc_img), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title(' Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv.filter2D(blur, -1, kernelx)
img_prewitty = cv.filter2D(blur, -1, kernely)
img_prewitt = img_prewittx + img_prewitty
plt.imshow(img_prewitt),plt.title('Prewitt')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Sobel
img_sobelx = cv.Sobel(blur,cv.CV_8U,1,0,ksize=5)
img_sobely = cv.Sobel(blur,cv.CV_8U,0,1,ksize=5)
img_sobel = img_sobelx + img_sobely
plt.imshow(img_sobel),plt.title('Sobel')
plt.xticks([]), plt.yticks([])
plt.show()
```
